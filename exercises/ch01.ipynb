```julia
using Plots
```

```julia
xvals = range(-10,10, length=1000)
```

```julia
#Real programming uses functions. So, I created the functions.
function create_norm_pdf(mu, sigma, xvals)
        #For some reason exp needed to be broadcast
        yvals = broadcast(exp,-0.5.*((xvals .- mu)./(sigma)).^2)./(sqrt(2*pi)*sigma)
        return yvals
    end
```

```julia
plot(xvals, create_norm_pdf(0,1, xvals), title="Standard Normal Distribution", label="Normal Distribution (0,1)")
```

```julia
plot(xvals, [create_norm_pdf(0,1, xvals) create_norm_pdf(0,2, xvals) create_norm_pdf(3,3, xvals)]
    ,title="Standard Normal Distribution"
    ,label = ["N(0,1)" "N(0,2)" "N(3,3)"]
)
```
